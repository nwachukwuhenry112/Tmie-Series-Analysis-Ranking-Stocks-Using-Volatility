```python
# AR Will be read from the PACF : Significant lags = 1,2,3,6,16,18,19,23,25,34, 35
# MA will be read from ACF : Significant lags = 1,2,3,6,8,16,18,19,23,25,34,35
# the lags that passes the standard error bands are significant

import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.arima_model import ARIMA

from scipy.stats.distributions import chi2 
from math import sqrt
import seaborn as sns
sns.set()
import warnings
warnings.filterwarnings("ignore")
```

```python
model_1 = ARIMA(train_int.int_ftse, order=(1,0,1))
results_1 = model_1.fit()
results_1.summary()

#  Did you apply ARIMA to initial ts or to the differenced one? If you applied for initial ts, 
# the d=1 seems to be good. But if you applied to differenced one, d=0 seems to be correct as you applied difference already
```

```python
model_2 = ARIMA(train_int.int_ftse, order=(3,1,2))
results_2 = model_2.fit()
results_2.summary()
```

```python
# to ensure model is not over fitting .  test using the loG Liklihood Ratio(LLR)
def LLR_test(mod_1, mod_2, DF=1):
    L1 = mod_1.fit().llf
    L2 = mod_2.fit().llf
    LR = (2*(L2-L1))
    p = chi2.sf(LR, DF).round(3)
    return p
```

```python
print("\nLLR test p-value = " + str(LLR_test(model_1, model_2,DF = 3)))
# order=(1,0,1)> 1+1 =2, order=(3,1,2)) > 3+2 = 5 SO Therefore, DF = 5 - 2 = 3
```

```python
# This implies the model is overfitting , SO use Model_1
## Analysing the Residuals
train_int['int_ftse_resid'] = results_1.resid
```
